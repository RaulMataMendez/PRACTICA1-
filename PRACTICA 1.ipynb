```
def Menu():
   print("                      CALCULADORA")
   print()
   print("PRESIONAR UN NUMERO SEGUN LA OPERACION QUE QUIERA RALIZAR")
   print("SUMA(1) RESTA(2) MULTIPLICACION(3) DIVISION (4)")
   print()

def Calculadora():
   
   op= int (input("Selecione Opcion: "))
   while (op >=1 and op <=4):
        No1 = float (input("Ingrese Numero: "))
        No2 = float (input("Ingrese Otro Numero: "))
        if (op==1):
            print ("La Suma es:", No1+No2)
            op = int (input("Selecione Opcion: "))
        elif (op==2):
            print("La Resta es:",No1-No2)
            op = int (input("Selecione Opcion: "))
        elif (op==3):
            print("La Multiplicacion es:",No1*No2)
            op = int (input("Selecione Opcion: "))
        elif (op==4):
            if (No2 !=0):
                print("La Division es:",No1/No2)
                op = int (input("Selecione Opcion: "))
            else:
              print("No es posible realizar esta opracion")
   else:
       print("Esta no es una operacion valida")
       print("Fin del pograma")
Menu()
Calculadora()
```

```

def salario(s):
   if (0<s<10000):
     print("Su salario menos impuestos es: ",s-.05*s)
     print("Usted gana menos de 10 mil: ",0<s<10000)
   elif (10000<=s<15000):
     print("Su salario menos impuestos es: ",s-.10*s)
     print("Usted gana entre 10 mil y 15 mil: ",10000<=s<15000)
   elif (15000<=s<=20000):
     print("Su salario menos impuestos es: ",s-.15*s)
     print("Usted gana entre 15 mil y 20 mil: ",15000<=s<=20000)
   elif (s>20000):
     print("Su salario menos impuestos es: ",s-.18*s)
     print("Usted gana mas de 20 mil: ",s>20000)
    
s=float (input("Introduzaca su salario: "))
salario(s)
```
